```python
# Solictiudes por Año
diesel_marino_solicitudes.groupby('Ano').sum()
```

```python
# Solicitudes Recibidas Distribuidas por Estado por Año
g = sns.factorplot(x="Estado", y="Solicitudes_Recibidas", col="Ano", 
                   data=diesel_marino_solicitudes,kind="bar", ci=None, aspect=2)
(g.set_axis_labels("", "Solicitues Recibidas").set_xticklabels(rotation=90))
```

```python
# Solicitudes Atendidas Distribuidas por Estado por Año
g = sns.factorplot(x="Estado", y="Activos_con_Consumos", col="Ano", 
                   data=diesel_marino_solicitudes,kind="bar", ci=None, aspect=2)
(g.set_axis_labels("", "Solicitues Atendidas").set_xticklabels(rotation=90))
```
